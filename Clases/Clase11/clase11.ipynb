 ## Módulos o Librerías
Python posee ciertos módulos con varias funciones y métodos incluidos, entre los más famosos estan: Sys, Os, Numpy, Matplotlib, Scipy, Sympy, Pandas, etc. A estos también se les llaman librerías. Antes de introducir módulos propios, introduciremos los siguientes:


1.   Sympy: Este módulo/librería permite la programación simbolica, es decir, el poder trabajar con variables matemáticas como tal, permitiendo la derivación, integración, suma y resta algebraicas y no solo aritmética. ([Documentación](https://docs.sympy.org/latest/index.html))
2.   Scipy: Este módulo permite trabajar programación científica de una forma más simple; brinda funciones y métodos para trabajar estadísitica, metodos numéricos, algebra lineal, etc. ([Documentación](https://docs.scipy.org/doc/scipy/reference/))



```
# Ejemplo Sympy
import sympy as sym
import math as m

# Se define la variable a utilizar
x = sym.symbols('x')

# Se define la función
f = sym.sin(x)

# Derivación de funciones
derivative_f = sym.diff(f,x,3)
print(derivative_f)

# Integración simbólica
integral_f = sym.integrate(f,(x,0,sym.pi))
print(integral_f)

# encontrar ceros de un polinomio
roots = sym.solve(x**3 - 6*x**2 + 11*x - 6,x)
print(roots)

# Simplificar una expresión
simple = sym.simplify((x**2 - 1)/(x + 1))
print(simple)
```

### Ejemplo:
Polinomios de Legendre:

Los polinomios de legrende son las funciones solución de la ecuación de legendre:

$$\frac{\text{d}}{\text{d} x} \left[ (1 - x^2) \frac{\text{d}}{\text{d} x} P_n (x) \right] + n(n+1) P_n (x)$$

Sin entrar en detalle, los polinomios de Legendre se definen como:
$$P_n (x) = \frac{1}{2^n n!} \frac{\text{d} ^n}{\text{d} x^n} [(x^2 - 1)^n]$$

Es normal preguntarnos, como entontrar el $n-$ésimo polinomio, o como se comporta gráficamente, o para el calculo de potenciales en física. Conesto, se introduce la idea en un programa simple.

```
# Polinomios de Legendre
import sympy as sym
import math as m

# Definimos la variable
x = sym.symbols('x')

# Creamos la lista donde se guardarán los Polinomios
Pn = []
n = int(input('Ingresa el número de polinomios deseado: '))

# Encontramos los primeros n polinomios de legendre
for i in range(n):
  c = 1 / (2**i * m.factorial(i))
  P = sym.diff(c*(x**2 - 1)**i,x,i)
  Pn.append(P)
  print(P)
  #print(sym.solve(c*P,x))

for i in range(len(Pn)):
  for j in range(len(Pn)):
    norm = sym.integrate(Pn[i]*Pn[j],(x,-1,1))
    print(norm, i + 1, j + 1)
```
