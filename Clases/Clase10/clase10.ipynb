## Método de Euler

El Método de Euler consiste en aproximar la solución de una ecuación diferencial (Con condiciones iniciales). El método se basa en definir la ecuación diferencial como:
  $$\frac{dy}{dx} = F(x,y)$$
Con condiciones iniciales:
  $$y(x_o) = y_o$$
Para calcular el valor de $y(x_n)$.


Por la definición de derivada, este método se basa en dividir el intervalo en cierta cantidad de partes e ir calculando en base a la pendiente de la solución en dicho "pedazo". Con lo que se tiene:
  $$y_{i+1} = y_{i} + F(x_i ,y_i) \Delta x$$
Donde, para el intervalo $[x_o, x_n]$ se tiene que $\Delta x = \frac{x_n - x}{n}$ ($n$ el número de subintervalos). \

Algoritmo:

1.   Tomar la ecuación diferencial como una función de dos variables.
2.   Iterar valuando la ecuación que determina el valor de $y$ $n$ veces.
3.   Crear una tabla en latex con los valores de cada iteración, con el formato siguiente:

|| i | $x_i$ | $y_i$ | $F(x_i,y_i)$ | $y_{i+1}$ ||

4.   Con la solución analítica de la ED, valuarla para el $x_n$, y encontrar el error porcentual de método.

$^\dagger$ El método de Ortonormalización de Gram-Schmidt se basa en encontrar los vectores ortonormales en base al primer vector de la base: \
$$e_1 = \frac{v_1}{||v_1||}$$
Y para los siguientes vectores:
$$e_j = \frac{v_j - \langle v_j,e_1 \rangle e_1 -\cdots - \langle v_j ,e_{j-1} \rangle e_{j-1}}{|| v_j - \langle v_j,e_1 \rangle e_1 -\cdots - \langle v_j ,e_{j-1} \rangle e_{j-1} ||}$$

$^\ddagger$ Para la matriz triangular superior se tiene lo siguiente: Si las filas o columnas de una matriz de $n\times n$ son linealmente independientes, entonces, existe una matriz ortogonal$^*$ $Q$ entonces lo siguiente es cierto:
$$A = QR$$
Donde $A$ es la matriz de análisis y $R$ es una matriz triangular superior. \

$^*$ Una matriz es ortogonal si se cumple lo siguiente:

*   Si sus vectores fila o vectores columna son ortonormales.
*   Si $QQ^T = I_n$.
*   Si $\det{Q} = \pm 1$.
