##if-else Block
 Un *if-else block* es una estructura de dos bloques que se basa en el valor de verdad de una condición. Si la condición es verdadera, el programa ejecuta el primer bloque del código. De lo contrario ejecutará el segundo bloque. Hay ocasiones en las que podemos emitir el bloque *else*. A continuación la estructura de un *if-else block* y su diagrama de flujo.

##elif 
Podemos combinar *if* con *else* para poner múltiples condiciones que sean mutuamente excluyentes. La estructura es la siguiente:

```
if condition1:
    <block of statements>
elif condition2:
    <block of statements>
elif condition3:
    <block of statements>
else:
    <block of statements>
<next statement>
```

Como ejemplo, haremos (e implementaremos en Python) un algoritmo para encontrar máximos y mínimos de una lista.  

Algoritmo para encontrar máximos:
1. Tener una lista con números aleatorios.
2. Asumir que el maximo se encuentra en la posición 0 de la lista.
3. Comparar el máximo que obtuvimos con el resto de números en la lista.
4. Seleccionar como al siguiente número más grande como nuevo máximo.
5. Hacer esto hasta encontrar al máximo absoluto de la lista.


```
#Inicializamos una lista
l=[1,5,3,-7,-1,-2312,-3,5,2,6,500]
sorted_l=[]
#Asumimos que el máximo y mínimo se encuentra en la posición 0
lmax=l[0]
lmin=l[0]

#Implementación de un ciclo for que haga el trabajo por nosotros

for i in range(len(l)):           #Explorar toda la lista
    if l[i]>=lmax:                #Condición para encontrar nuevos máximos
        lmax = l[i]               #Definir el nuevo máximo

    elif l[i]<=lmin:              #Condición para encontrar nuevos mínimos  
        lmin = l[i]               #Definir el nuevo mínimo

#Imprimimos los resultados obtenidos
print('El máximo de la lista es', lmax)                        
print('El mínimo de la lista es', lmin)
```
